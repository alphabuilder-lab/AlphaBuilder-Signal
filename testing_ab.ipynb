```python
# stock_dashboard.ipynb
import yfinance as yf
import ipywidgets as widgets
from IPython.display import display
import matplotlib.pyplot as plt
import pandas as pd

# Dropdown for tickers
tickers = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'META']
ticker_dropdown = widgets.Dropdown(options=tickers, description='Ticker:')
start_date_picker = widgets.DatePicker(description='Start Date', value=pd.to_datetime('2023-01-01'))
end_date_picker = widgets.DatePicker(description='End Date', value=pd.to_datetime('2023-12-31'))
button = widgets.Button(description="Plot")

output = widgets.Output()

def plot_stock_data(b):
    output.clear_output()
    ticker = ticker_dropdown.value
    start = start_date_picker.value
    end = end_date_picker.value
    data = yf.download(ticker, start=start, end=end)
    
    with output:
        if data.empty:
            print("No data found for the selected ticker/date range")
        else:
            plt.figure(figsize=(10, 5))
            plt.plot(data['Close'], label=f'{ticker} Close')
            plt.title(f'{ticker} Closing Prices')
            plt.xlabel('Date')
            plt.ylabel('Price')
            plt.legend()
            plt.show()

button.on_click(plot_stock_data)

display(widgets.VBox([ticker_dropdown, start_date_picker, end_date_picker, button, output]))
```
